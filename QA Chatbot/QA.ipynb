Import All libraries

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np
```

Load Data Set

```python
df=pd.read_csv('QA Chat Bot Dataset.csv')
```

#faq and ans are separated

```python
faq=df.drop("Answer",axis=1)
ans=df["Answer"]
```

```python
faq.head
```

```python
ans.head()
```

Apply text vectorizer and lower case the text at the same time to generate the same vector for the same data present in lowercase or upper case

```python
vec=TfidfVectorizer(encoding='utf-8',lowercase=True)
```

```python
faq_vec=vec.fit_transform(faq["Question"])
```

```python
faq_vec.shape
```

```python
faq_vec.toarray()
```

Implement FAQ answer matching logic

```python
def get_faq_answer(user_query: str, threshold=0.4):
    query = vec.transform([user_query])
    similarity = cosine_similarity(query, faq_vec).flatten()

    best_match = similarity.argmax()
    score = similarity[best_match]

    if score > threshold:
        return ans.iloc[best_match]

    else:
      return "sorry sir! I don't have any knowledge about it. please ask anything about services"
```

```python
get_faq_answer("What is your return policy?")
```
